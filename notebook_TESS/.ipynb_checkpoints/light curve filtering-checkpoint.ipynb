```python
import batman
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Our parameters are here https://exofop.ipac.caltech.edu/tess/
# Initialize transit parameters
params = batman.TransitParams()
params.t0 = 0.0                      # time of inferior conjunction
params.per = 1.0                     # orbital period
params.rp = 0.1                      # planet radius (in units of stellar radii)
params.a = 15.0                      # semi-major axis (in units of stellar radii)
params.inc = 87.0                    # orbital inclination (in degrees)
params.ecc = 0.0                     # eccentricity
params.w = 90.0                      # longitude of periastron (in degrees)
params.u = [0.1, 0.5]                # limb darkening coefficients [u1, u2]
params.limb_dark = "quadratic"       # limb darkening model

# Time array
t = np.linspace(-0.025, 0.025, 1000)

# Initialize model and calculate light curve
m = batman.TransitModel(params, t)
flux = m.light_curve(params)

# Plot the light curve
plt.figure(figsize=(6, 4))
plt.plot(t, flux, label='Light Curve')
plt.xlabel('Time from central transit [d]')
plt.ylabel('Relative flux')
plt.title('Transit Light Curve')
plt.legend()
plt.show()

# we are going to use https://github.com/hpparvi/ldtk 
```

```python
# check the filter that was used in Asiago here http://svo2.cab.inta-csic.es/theory/fps/
```

```python
from ldtk import SVOFilter, LDPSetCreator
SVOFilter.shortcuts
```

```python
sloan_g = SVOFilter('sdss_g') # pay attention that here red by convetion
sloan_r = SVOFilter('sdss_r') # pay attention that here red
sloan_rprime = SVOFilter('SLOAN/SDSS.rprime_filter')
tess_filter = SVOFilter('tess')

fig, ax = plt.subplots(figsize=(6,4))
sloan_g.plot(ax=ax)
sloan_r.plot(ax=ax)
sloan_rprime.plot(ax=ax)
tess_filter.plot(ax=ax)
plt.show()
```

```python
filters = [sloan_r, tess_filter]
# value from temperature and assosiated arror
sc = LDPSetCreator(teff=(3652, 50), logg=(4.78, 0.12), z=(0.2, 0.10), filters=filters)

ps = sc.create_profiles(nsamples=2000)
ps.resample_linear_z(100)

ps.set_uncertainty_multiplier(10)
qm, qe = ps.coeffs_qd(do_mc=True, n_mc_samples=10000)
#look at the paper to understand the theory behind
chains = np.array(ps._samples['qd'])
```

```python
print(np.shape(chains))
```

```python
u1_sloan_r_chains = chains[0,:,0]
u2_sloan_r_chains = chains[0,:,1]

u1_tess_chains = chains[1,:,0]
u2_tess_chains = chains[1,:,2]

print('Sloan r LD coefficients u1 = {0:.4f} \pm {1:.4f}  u2 = {2:.4f} \pm {3:.4f}'.format(
    np.median(u1_sloan_r_chains, np.std(u1_sloan_r_chains)),
    np.median(u2_sloan_r_chains, np.std(u2_sloan_r_chains)),
))

print('Sloan  LD coefficients u1 = {0:.4f} \pm {1:.4f}  u2 = {2:.4f} \pm {3:.4f}'.format(
    np.median(u1_tess_chains, np.std(u1_tess_chains)),
    np.median(u2_tess_chains, np.std(u2_tess_chains)),
))

plt.figure(figsize=(4,4))
plt.title('Sloan r LD coefficients')
plt.scatter(u1_sloan_r_chains, u2_sloan_r_chains)
plt.xlabel('u1')
plt.ylabel('u2')
plt.show()

plt.figure(figsize=(4,4))
plt.title('TESS r LD coefficients')
plt.scatter(u1_tess_chains, u2_tess_chains)
plt.xlabel('u1')
plt.ylabel('u2')
plt.show()
```
