```python
import numpy as np
import matplotlib.pyplot as plt 
#%matplotlib notebook
%matplotlib widget 
from astropy.io import fits
from astropy.wcs import WCS
```

```python
sector44_tpf = 'tess2020078014623-s0023-0000000198108326-0177-s_tp.fits'
sector44_lcf = 'tess2020078014623-s0023-0000000198108326-0177-s_lc.fits'
```

```python
fits.info(sector44_tpf)
```

```python
tpf_hdu = fits.open(sector44_tpf)
#tpf_hdu[0].header
print(tpf_hdu[0].data)

```

```python
tpf_hdu[1].columns
```

```python
tpf_data = tpf_hdu[1].data
# print(tpf_data) 

sector44_bjdtdb = tpf_data['TIME'] + 2457000.0
# print(sector44_bjdtdb)

first_image = tpf_data['FLUX'][0]
# print(np.shape(tpf_data['FLUX']))

wcs = WCS(tpf_hdu[2])
```

```python
fig = plt.figure(figsize=(7,7))

fig.add_subplot(111, projection = wcs)
plt.imshow(first_image, origin='lower', cmap=plt.colormaps['viridis'])
#    , vmax=4100
plt.xlabel(r'RS', fontsize=14)
plt.ylabel(r'DC', fontsize=14)
plt.grid(axis='both', color='white', ls='solid')
plt.show()
```
